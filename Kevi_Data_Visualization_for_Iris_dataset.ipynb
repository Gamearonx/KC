```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
file_path = "/content/Iris.csv"
df = pd.read_csv(file_path)
# Display the first few rows of the dataset
print(df.head)
```

```python
### Question 1: What is the distribution of sepal lengths and widths among the Iris species?
# Boxplot to compare the sepal lengths and widths across species
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Sepal Length Distribution by Species')
plt.show()

sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Sepal Width Distribution by Species')
plt.show()
```

```python
### Inference
# Iris Setosa has the smallest sepal sizes, while Iris Virginica has the biggest, and Iris Versicolor is somewhere in the middle.
```

```python
### Question 2: How does the petal length vary across different Iris species?
# Plotting petal length variation accross Iris species
sns.boxplot(x='PetalLengthCm', hue='Species', data=df)
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
### Inference
# The boxplot shows that Iris Setosa has the smallest petals, Iris Virginica has the biggest, and Iris Versicolor has petals that are in between.
```

```python
### Question 3 : Is there a relationship between sepal length and petal length?
# Scatter plot for sepal length vs petal length
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=df)
plt.title('Relationship Between Sepal Length and Petal Length')
plt.show()
```

```python
### Inference
# The scatter plot shows that there is a positive relationship between sepal length and petal length: as sepal length increases, petal length tends to increase as well.
```

```python
### Question 4 : How does the petal width vary across different Iris species?
# Plotting petal length variation accross Iris species
sns.boxplot(x='PetalWidthCm', hue='Species', data=df)
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
### Inferene:
# The boxplot shows that Iris Setosa has the narrowest petal widths, while Iris Virginica has the widest, with Iris Versicolor falling in between.
```

```python
## Question 5: Are there any patterns observable in the pairwise relationships between features
#sepal length, sepal width, petal length, petal width?
# Pairplot to visualize pairwise relationships between features
# Select relevant features
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
sns.pairplot(df, hue='Species')
plt.suptitle('Pairwise Relationships Between Features')
plt.show()
```

```python
### Inference
# The pairplot shows that **Iris Setosa** is easy to separate from the other two species, while Iris Versicolor and Iris Virginica are a bit mixed but still look different from each other.
```

```python
### Question 6: Can we distinguish between Iris species based on sepal and petal measurements?
# Violinplot to distinguish between Iris species based on sepal and petal measurements
# Plot violin plots for all features
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
plt.figure(figsize=(8, 6))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.violinplot(x='Species', y=feature, data=df, palette='bright')
    plt.title(f'{feature} Distribution by Species')

plt.tight_layout()
plt.show()
```

```python
### Inference
# Iris setosa is easy to recognize because it has smaller petals and sepals. Iris versicolor and Iris virginica are harder to tell apart, but their petal sizes can help you figure out which one is which.
```
