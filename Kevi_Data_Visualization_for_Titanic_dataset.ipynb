```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
file_path = "/content/train.csv"
df = pd.read_csv(file_path)
# Display the first few rows
print(df.head())
```

```python
### Question 1: What is the distribution of passenger ages on the Titanic?
# Plotting distribution of passenger ages on the Titanic
sns.histplot(df['Age'], bins=30, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
### Inference
# Most Titanic passengers were young adults, with fewer older individuals and some children.
```

```python
### Question 2: How does the survival rate vary based on passenger class (Pclass)?
# Plotting survival rate variation based on passenger class
sns.boxplot(x='Pclass', y='Age', data=df, palette='bright')
plt.title('Age vs Pclass')
plt.show()
```

```python
### Inference
# Being in a higher class really helped people stay safe on the Titanic.
```

```python
### Question 3: Is there a relationship between the fare paid and the likelihood of survival?
# Plotting relationship between fare paid and survival
sns.boxplot(x='Survived', y='Fare', data=df, palette='pastel')
plt.title('Fare Paid vs. Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare Paid')
plt.show()
```

```python
### Inference
# Passengers who paid higher fares had a greater chance of surviving the Titanic disaster.
```

```python
### Question 4: How does the survival rate differ between male and female passengers?
# Plotting survival rate between and femal passengers
sns.barplot(x='Sex', y='Survived', data=df, ci=None, palette='bright')
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.show()
```

```python
### Inference
# Female passengers had a significantly higher survival rate than male passengers on the Titanic.
```

```python
### Question 5: What impact does the number of siblings/spouses (SibSp) or parents/children
Parch aboard have on survival?
# Plotting parch onboard survival between siblings/spouses (SibSp) or parents/children
sns.countplot(x='SibSp', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Siblings/Spouses Aboard')
plt.show()
```

```python
### Inference
# Passengers with 12 family members had better survival chances, while those traveling alone or with larger families had lower rates.
```

```python
### Question 6: Are there any significant differences in survival rates among passengers embarked
from different ports?
# Plotting survival by embarkation port
sns.countplot(x='Embarked', hue='Survived', data=df, palette='Set2')
plt.title('Survival by Embarkation Port')
plt.show()

# Calculate survival rate by embarkation port
survival_embarked = df.groupby('Embarked')['Survived'].mean() * 100
print("Survival Rate by Embarkation Port:\n", survival_embarked)
```

```python
### Inference
# Passengers who embarked from Cherbourg (C) had the highest survival rate, followed by Southampton (S) and Queenstown (Q), indicating location-based survival disparities.
```
